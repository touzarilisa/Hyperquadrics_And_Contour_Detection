```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from prgm import fonction_LM,fonction_HQ
from initialise_coefHQ import initialise_coefHQ
```

## Test de la méthode de Levenberg-Marquardt sur différentes données  test1 "hq1b1"                       

```python
                         
with open('hq1b1.csv', 'r') as f:
    data = np.array([[float(e) for e in row] for row in csv.reader(f, delimiter = ',', quoting = csv.QUOTE_NONE)])
    
x_pts = data[0]    
y_pts = data[1]
v = 10 ** 8
k1 = 10
k2 = 10
nmax = 200 #nombre d'iteration
precision = 0.000001
smax1 = 140 #diametre de la surface maximal
smin1 = 35 #diametre de la surface minimal
NhT = [3,4,5,6] #les coefficient utilise

for Nh in NhT :
    
    l = initialise_coefHQ(x_pts, y_pts, Nh) # les coefs A B C
    l = l[:, 0:3]
    lm =fonction_LM(x_pts, y_pts, l, v, k1, k2, smax1, smin1, precision, nmax)
    #utilisation de la methode de Levenberg - Marquardt 
    a = lm[:, 0]# Recuperation des 
    b = lm[:, 1]# coefs A 
    c = lm[:, 2]# B et C
    gamma = []
    
    for i in range (Nh):
        gamma.append(4) #definition du vecteur gamma

    xmin, xmax, nx = 0, 300, 41
    ymin, ymax, ny = 0, 300, 41
    
    # Discrétisation du domaine de tracé
    x1d = np.linspace(xmin, xmax, nx)
    y1d = np.linspace(ymin, ymax, ny)
    x2d, y2d = np.meshgrid(x1d, y1d)
    plt.figure(figsize = (10,15))
    plt.subplot(2,2,(Nh-2))
    plt.grid()
    
    # Tracé de HQ et des droites englobantes
    plt.contour(x2d, y2d, fonction_HQ(x2d, y2d ,a, b ,c ,gamma, Nh), colors = ['r'])
    plt.scatter(x_pts,y_pts,marker = '+')
    plt.title('Isovaleurs POUR nh = {}'.format(Nh))
    plt.xlabel('Valeurs de x')
    plt.ylabel('Valeurs de y')
    plt.grid()
    plt.axis('square')
```

## COMMENTAIRES :

**on a choisit des valeurs de nh tel que nh = [3,4,5,6].**

  **-on remarque que hq1b1 a plus de points que les autres fichiers  c'est pour ça on obtient une courbe plus 
      précise qui passe par plus de points.**
    
  **-on constate que lorsque nh augmente  la courbe passe mieux par le nuage de points.**
    
   **-on remarque que pour nh=3 l'algorithme diverge  et pour nh=4 la courbe ne passe pas prés de tous les points 
     et par contre pour nh=5,6 la courbe passe bien prés de tous les points .**
